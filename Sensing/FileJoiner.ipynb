```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.compat import StringIO
from datetime import datetime
from tqdm import tqdm
```

```python

filepath = "./Data/2020-09-20/"

output = pd.DataFrame(columns = ["Year","Month","Day","Hour","Minute","Second","Record","U (m s^-1)","V (m s^-1)","W (m s^-1)","TSonic (K)","RH (%)","T_HMP60 (C)"])

dict={}
for i in tqdm(range(1440)):
    outputsize = len(dict)
    
    df = pd.read_table(filepath+"Younng10Hz"+str(i)+".dat",
                       sep=",",
                      skiprows=[0,2,3])

    for j in range(len(df)):
        if "." in df['TIMESTAMP'][j]:
            date = datetime.strptime(df['TIMESTAMP'][j], '%Y-%m-%d %H:%M:%S.%f')
            #print(date)
        else:
            date = datetime.strptime(df['TIMESTAMP'][j], '%Y-%m-%d %H:%M:%S')
            
        dict[outputsize+j]={"Year":date.year,"Month":date.month,"Day":date.day,"Hour":date.hour,"Minute":date.minute,"Second":date.second+date.microsecond/1000000.,"Record":df['RECORD'][j],"U (m s^-1)":df['U'][j],"V (m s^-1)":df['V'][j],"W (m s^-1)":df['W'][j],"TSonic (K)":df['TSonic'][j],"RH (%)":df['RH'][j],"T_HMP60 (C)":df['T_HMP60'][j]}
            
            
        #output.loc[outputsize+j,output.columns]=[date.year,date.month,date.day,date.hour,date.minute,date.second+date.microsecond,df['RECORD'][j],df['U'][j],df['V'][j],df['W'][j],df['TSonic'][j],df['RH'][j],df['T_HMP60'][j]]
        #output = output.append(pd.Series([date.year,date.month,date.day,date.hour,date.minute,date.second+date.microsecond/1000000.,df['RECORD'][j],df['U'][j],df['V'][j],df['W'][j],df['TSonic'][j],df['RH'][j],df['T_HMP60'][j]], index=output.columns ), ignore_index=True)
    
output = pd.DataFrame.from_dict(dict, "index")


output.to_csv('testoutput.csv',index=False)






```
