```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

fileName = './Raw Data/noControlCircle5s_raw.xlsx'

def dist(x1, y1, x2, y2, r):  
    return np.sqrt(((x2 - x1)** 2) + ((y2 - y1)** 2)) - r;

mpl.rcParams['figure.dpi'] = 300
```

```python
data = pd.read_excel(fileName)
data.columns=(['timeglobal','timenow','rb1x','rb1y','rb1z','rb2x','rb2y','rb2z','rb3x','rb3y','rb3z','rb4x','rb4y','rb4z','rb5x','rb5y','rb5z','u1','u2','u3','u4'])

#get rid of zeros at the end
dataToDrop = []
for i in range(len(data['timeglobal'])):
    if data['timeglobal'][i]==0:
        dataToDrop = np.append(dataToDrop,i)
    
data = data.drop(dataToDrop)
#print(data)

#print(data)
```

```python
plt.plot(data['rb5x'][800:2400],data['rb5y'][800:2400])
plt.xlabel('Y position (mm)')
plt.ylabel('X position (mm)')
plt.title('Uncontrolled Circular Path')
plt.gca().set_aspect('equal')

print(max(data['rb5y'][500:])-min(data['rb5y'][500:]))
print(max(data['rb5x'][500:])-min(data['rb5x'][500:]))
xc = (np.average(data['rb5y'][500:]))
yc = (np.average(data['rb5x'][500:]))

error = []
for i in range(len(data['rb5x'][500:2500])):
    error = np.append(error,(dist(xc,yc,data['rb5x'][i+500],data['rb5y'][i+500],480))**2)

print(np.sqrt(sum(error)/len(error)))

sortederror = np.sort(error)
print(np.sqrt(sum(error[int(len(error)*0.05):int(len(error)*0.1)])/int(len(error)*0.05)))

plt.show()

plt.plot(np.sqrt(error))
#print(error)
```
